Run the following load command to get the 2-Dim iris dataset. Do the following operations on it.

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
```

Drop rows that contain a missing value. 

```python
iris_2d_drop=iris_2d[~np.isnan(iris_2d).any(axis=1),:]
```

Find the correlation between first two columns.

```python
corr=np.corrcoef(iris_2d_drop[:,0],iris_2d_drop[:,1],rowvar=False)
corr[0,1]
```

Find the mean,median and standard deviation of column 3.

```python
mean=np.mean(iris_2d_drop[:,2])
median=np.median(iris_2d_drop[:,2])
std=np.std(iris_2d_drop[:,2])
print(mean, median, std)
```

Create a new column for volume in iris_2d, where volume is (pi x petallength x sepal_length^2)/3

```python
petallength=iris_2d_drop[:,0].astype('float')
sepal_length=iris_2d_drop[:,2].astype('float')
vol=(np.pi * petallength * (sepal_length**2))/3
vol = vol[:, np.newaxis]
iris_2d_new = np.append(iris_2d_drop, vol ,axis=1)
iris_2d_new
```
